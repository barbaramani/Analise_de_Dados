```python
import pandas as pd

gateway = pd.read_pickle(r'files/gateway.pkl')
measurements = pd.read_pickle(r'files/measurements.pkl')
sensors = pd.read_pickle(r'files/sensors.pkl')
gateway
```

```python
gateway['name'] = gateway['name'].str.upper().str.replace('_',' 0')
gateway
```

```python
gateway_df = pd.DataFrame(gateway)
gateway_df  = gateway_df.drop_duplicates()
gateway_df
```

```python
df_gateway = pd.DataFrame(gateway)
df_sensors = pd.DataFrame(sensors)
```

```python
df_merged = pd.merge(df_sensors, df_gateway, left_on='gatewayId', right_on='id', how='right')
```

```python
sensor_count_per_gateway = df_merged.groupby(['id_y', 'name_y'])['id_x'].count().reset_index()
sensor_count_per_gateway.columns = ['gatewayId', 'gatewayName', 'sensorCount']

sensor_count_per_gateway
```
